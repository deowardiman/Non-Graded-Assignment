```python
import sqlite3
import pandas as pd
import csv
```

Ada 3 fungsi untuk membantu:
- Yang buat connection dgn database
- Yang execute perintah SQL
- Yang milih dan nampilin dari database

```python
def create_connection(path):
    connection = sqlite3.connect(path)
    return connection
```

```python
def execute_query(connection, query):
    cursor = connection.cursor()
    cursor.execute(query)
    connection.commit()
    print('Executed')
```

```python
def select_query(conn, query):
    cur = conn.cursor()
    cur.execute(query)
    
    rows = cur.fetchall()
    
    for row in rows:
        print(row)
```
